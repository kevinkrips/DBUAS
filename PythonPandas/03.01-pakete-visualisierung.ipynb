# Visualisierung von Daten in Python

**Literaturempfehlung:**

**Pakete:**

(1) [Matplotlib](https://matplotlib.org/) ist das wichtigste Pakete für die Visualisierung von Daten in Python. Alle anderen Pakete mit denen wir arbeiten werden bauen auf Matplotlib auf.

(2) [Seaborn](https://seaborn.pydata.org/) bietet eine umfassende Sammlung von sogenannten _high-level interfaces_ für die Visualisierung von Daten basierend auf Matplotlib an. Mit _high-level interface_ ist gemeint, dass die Definition dessen, wie eine Visualisierung am Ende aussehen soll, deutlich abstrakter ist. Während Matplotlib sich sehr stark daran orientiert, wie welche Linien gezeichnet werden sollen, konzentriert sich Seaborn mehr auf die Daten und welche Zusammenhänge aus den Daten ersichtlich werden sollen. Die folgende Unteseite bietet einen guten Überblick über die verfügbaren Diagramtypen: https://seaborn.pydata.org/api.html

(3) [Pandas](https://pandas.pydata.org/) bietet selbst eine Reihe von Visualisierungsoptionen, wiederum basierend auf Matplotlib. Für die meisten Anwendungsfälle ist es einfacher, mit Seaborn zu arbeiten. Den größten Vorteil in der arbeit mit Pandas sehe ich darin, das auch aggregierte Tabellen (beispielsweise crosstab) direkt visualisiert werden können.

Darüber hinaus gibt es noch viele andere Pakete für die Visualisierung von Daten, beispielsweise [ggplot](http://ggplot.yhathq.com/), [Altair](https://altair-viz.github.io/) oder [Bokeh](https://bokeh.org/). Ihr seid natürlich herzlich eingeladen, diese Pakete auszuprobieren. Für die Bearbeitung der Übungen und später der Studienarbeit möchte ich Euch aber bitten, nur mir Matplotlib, Seaborn und Pandas zu arbeiten um eine Vergleichbarkeit der Leistungen zu ermöglichen.

## Matplotlib in Jupyter

Abgesehen davon, dass wir nun matplotlib als Paket laden, aktivieren wir mit `%matplotlib inline` die Darstellung von Visualisierungen direkt im Jupyter Notebook.

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips["total_bill"].plot()
```

```python
tips["total_bill"].plot.line()
```

```python
tips["total_bill"].plot.box()
```

```python
tips["total_bill"].plot.hist()
```

## Exkurs: Aggregierte Tabellen

Wenn wir Kreuztabellen und ähnliche aggregierte Übersichten in Pandas erzeugen, so sind diese eigentlich auch immer entweder eine Series oder ein DataFrame. Entsprechend können wir auf Ergebnistabellen auch alle Befehle anwenden, die wir beim Datensatz selbst nutzen.

```python
day_count = tips["day"].value_counts()
day_count
```

```python
day_count.plot.bar()
```
