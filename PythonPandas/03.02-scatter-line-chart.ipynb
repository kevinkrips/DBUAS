# Scatter plot and line chart

Wie angekündigt, arbeiten wir bei der Visualisierung von Daten mit drei Paketen: Matplotlib, Pandas und Seaborn. In diesem Notebook zeige ich Euch Beispiele für _scatter plots_ (Punktwolken) und _line charts_ (Liniendiagramme) mithilfe aller drei Pakete.

**Dokumentation der Pakete:**

(1) Matplotlib: [Tutorials](https://matplotlib.org/tutorials/index.html#) und [Gallery mit Beispielen](https://matplotlib.org/gallery/index.html)

(2) Pandas: [Plot-Dokumentation](https://pandas.pydata.org/pandas-docs/stable/reference/frame.html#plotting)

(3) Seaborn: [API Reference](https://seaborn.pydata.org/api.html)

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

## Scatter Plot: Matplotlib

```python
plt.scatter(tips["total_bill"], tips["tip"])
```

```python
plt.scatter(tips["total_bill"], tips["tip"], alpha=0.5)
```

## Scatter Plot: Pandas

```python
tips.plot.scatter("total_bill", "tip")
```

## Scatter Plot: Seaborn

```python
sns.scatterplot("total_bill", "tip", data=tips)
```

```python
sns.lmplot("total_bill", "tip", data=tips)
```

```python
sns.lmplot("total_bill", "tip", hue="smoker", data=tips)
```

```python
sns.lmplot("total_bill", "tip", col="smoker", data=tips)
```

## Line Chart: Matplotlib (Beispiel 1)

```python
tips["total_bill"].plot()
```

```python
total_bill = tips["total_bill"].sort_values()
total_bill[0:9]
```

```python
plt.plot(total_bill)
```

```python
total_bill = list(total_bill)
total_bill[0:9]
```

```python
plt.plot(total_bill)
```

## Line Chart: Matplotlib (Beispiel 2)

```python
percentage = (tips["tip"] / tips["total_bill"] * 100).round(2)
percentage = list(percentage.sort_values())
percentage[0:9]
```

```python
plt.plot(percentage)
```

```python
plt.scatter(range(len(percentage)), percentage)
```

```python
pd.Series(percentage).describe()
```

```python
plt.plot(percentage)
plt.scatter(range(len(percentage)), percentage)
```

## Line Chart: Seaborn

```python
sns.lineplot(range(len(percentage)), percentage)
```
